```python
# class practice using goodreads 

import urllib 
from bs4 import BeautifulSoup  

def get_search_url_goodreads(search_term):
    url = 'https://www.goodreads.com/search?query=%s' % (str(search_term))
    return url 

def get_search_url_bgg(search_term):
    url = 'https://boardgamegeek.com/geeksearch.php?action=search&objecttype=boardgame&q=%s' % (str(search_term))
    return url

def get_page_content(url):
    req = urllib.request.Request(url)
    with urllib.request.urlopen(req) as response:
        return response.read()
```
